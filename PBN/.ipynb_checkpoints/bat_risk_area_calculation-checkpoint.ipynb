Step 1: Visualize the shapefiles, inspect what I'm working with

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Read input
aoi = gpd.read_file("aoi.shp")
forested_areas = gpd.read_file("forested_areas.shp")
```

```python
# Display params
lyr = aoi.plot(color="white", edgecolor="black")
forested_areas.plot(ax=lyr, color="red")
plt.show()
```

Step 2: Calculate forested areas in acres

```python
# Estimate the UTM Zone, and project into estimated UTM Zone
forested_areas.estimate_utm_crs()
```

```python
forested_areas_utm = forested_areas.to_crs(32615)

# 1 sq. meter ≈ 0.000247105 acres (divide by 4047)
forested_areas_utm['acreage'] = forested_areas_utm.area / 4047
forested_areas_utm.crs
```

Step 3: Find suitable bat habitats

- Forested area – a contiguous forest polygon, i.e., a data row in the forested_areas.shp
- Suitable bat habitat – all large (>= 10 acre) forested areas, plus any small (< 10 acre)
forested area within a “flight-radius” of a large forested area
- Flight-radius – 1000 feet, the maximum distance from their habitat that bats will fly to
conduct their day-to-day business, as defined by USFWS

```python
less_than_ten_acres = forested_areas_utm[forested_areas_utm['acreage'] < 10]
greater_than_ten_acres = forested_areas_utm[forested_areas_utm['acreage'] > 10]
```

```python
# Find all forested areas less than 10 acres within the "flight radius" (1000 feet) 
# of forested areas greater than 10 acres (suitable bat habitats)
# 
# 1 foot = 0.3048 meters (divide by 3.281)
flight_radius_meters = 1000 / 3.281
```

```python
# Buffer large forested areas by fliht radius and union to one polygon
greater_than_ten_acres_buff = greater_than_ten_acres.buffer(flight_radius_meters)
greater_than_ten_acres_buff_union = greater_than_ten_acres_buff.geometry.unary_union
```

```python
# Calculate small areas within union of buffer
suitable_small_areas = less_than_ten_acres[less_than_ten_acres.within(greater_than_ten_acres_buff_union)]
suitable_small_areas
```

```python
# Concatenate the GeoDataFrames of large forested areas and suitable small areas
gdf_list = [greater_than_ten_acres, suitable_small_areas]
suitable_bat_habitat = gpd.GeoDataFrame(pd.concat(gdf_list, ignore_index=True), crs=gdf_list[0].crs)
suitable_bat_habitat
```

Step 4: Calculate Bat risk area

• Rotor radius – 88 meters. Although turbines are modeled as GIS center points, their rotor
blades extend their operational geographic footprints

• Project area of interest – polygon boundary, within which a wind farm will be constructed

• Bat risk area – a polygon area, within a project area of interest, where placing a turbine
(center point) would put bats at risk

```python
# Project the area of interest (aoi) to utm
aoi_utm = aoi.to_crs(32615)
```

```python
# Some invalid polygons, buffer of 0 fixes polygon errors
suitable_bat_habitat = suitable_bat_habitat.buffer(0)
```

```python
# Clip suitable bat habitat to the AOI. Output is bat risk area
suitable_bat_habitat_clip = suitable_bat_habitat.clip(aoi_utm)  
bat_risk_area = suitable_bat_habitat_clip
```

```python
# Display bat risk area
bat_risk_area.plot()
plt.show()
```

```python
# Export to shapefile
bat_risk_area.to_file("bat_risk_area.shp")
```
